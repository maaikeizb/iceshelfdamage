````python
from math import sin

# with a loop:
x = np.linspace(0,2*np.pi,10)
y = []
for a in x:
    y.append(sin(a))
    print(sin(a))
    
    
# numpy is not clean: you have to think for yourself if it is a single number or an array 
# (for which np.sin gives different output)
y = np.sin(x)
y1 = np.sin(2)
print(y)
print(y1)    

# more declarative language: better readable!
y = [sin(a) for a in x]
print(y)

# always a tradeoff between convenient numpy (for either float or array) or better readable
````
